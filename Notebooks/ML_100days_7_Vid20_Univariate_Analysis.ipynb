```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('matches.csv')
df.sample(10)
```

# **On Categorical data**

```python
sns.countplot(df['toss_decision']) #countplot
```

```python
df.value_counts('toss_decision')
```

```python
df['toss_decision'].plot(kind='bar')
```

```python
df['toss_decision'].value_counts().plot(kind='bar') #IMPORTANTTTTTTTTTTTTTTT: MUST ADD 'value_counts()'
```

```python
df['toss_decision'].value_counts().plot(kind='pie', autopct='%.2f') #IMPORTANT: 'autopct' uses for plotting percentage along with ONLY 'PIECHART', I think. See next cell for 'autopct' on bar
```

```python
df['toss_decision'].value_counts().plot(kind='bar', autopct='%.2f')
```

```python
df['toss_decision'].value_counts().skew()
```

# **On Numeric data**



> ****FIRST LESSON* *** ***WHILE WORKING WITH UNIVARIATE NUMERIC DATA IS TO CREATE A HISTOGRAM AT VERY FIRST***



```python
import matplotlib.pyplot as plt
plt.hist(df['win_by_wickets']) #histogram
```

```python
df['win_by_wickets'].value_counts()
```

```python
plt.hist(df['win_by_wickets'], bins=50) #BEST APPROACH is to provide a logical larger bins numbers
```

```python
sns.distplot(df['win_by_runs']) #deprecated function
```

```python
sns.distplot(df['win_by_runs'], hist=False)
```

```python
sns.distplot(df['win_by_wickets']==10)
```

```python
sns.distplot(df['win_by_wickets']==10, hist=False)
```

```python
sns.distplot(df[df['win_by_wickets']==10]['season'], hist=False)
```

```python
sns.distplot(df['win_by_runs'], hist=False)
sns.distplot(df[df['win_by_wickets']==10]['season'], hist=False)
```

```python
sns.distplot(df[df['win_by_runs']==2]['season'])
sns.distplot(df[df['win_by_wickets']==10]['season'])
```

```python
sns.distplot(df[df['win_by_runs']==2]['season'], hist=False)
sns.distplot(df[df['win_by_wickets']==10]['season'], hist=False)
```

```python
sns.displot(df['win_by_wickets']) #displot
```

```python
sns.histplot(df['win_by_wickets']) #histplot
```

```python
df['win_by_wickets'].skew() #skew is an statistical concept which measures skewness of data. 0 means skew is perfect, positive skew means the data is positively skewed to the right side, and negative skew means the data is negatively skewed to the left side
```
