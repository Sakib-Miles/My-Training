

> ***Now we will create a loop to iterate all 483 pages***



```python
#Making an empty database first
df = pd.DataFrame()
df
```

```python
for i in range(0, 484):
  response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(i)) #replaced a dictionary in the place of 1 and iterating page numbers
  temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
  df = pd.concat([df, temp_df],ignore_index=True) #concatenating one DataFrame after another ('MOST IMPORTANT: CONCATENATION "NEEDS A LIST AS FIRST ARGUMENT" TO UNDERSTAND THE ORDER OF ITERATION') and ignoring once iterated DataFrames
```

```python
for i in range(1, 484): #"MUST START FROM 1"
  response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(i))
  temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
  df = pd.concat([df, temp_df],ignore_index=True)

df
```
